```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sk
from sklearn.linear_model import LinearRegression
```

```python
#Correlation coeficient 
def index_corr_coef(y, y_pred):
    #Error sum of squares
    ess = np.sum(np.square(y - y_pred))
    #total sum of squares 
    tss = np.sum(np.square(y - np.mean(y)))
    
    index = np.sqrt(1 - ess/tss)
    return index 

#Mean squared error 
def meanse(y, y_pred):
    return np.mean(np.square(y - y_pred))
```

```python
#Linear regression model 
np.random.seed(0)
x = np.arange(-5.0, 5.0, 0.1)
y = 3*x + np.random.rand(100)
y_noise = 2 * np.random.normal(size=x.size)
y = y + y_noise

# transforming the data to include another axis
x = x[:, np.newaxis]
y = y[:, np.newaxis]

model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)

mse = meanse(y, y_pred)
corr = index_corr_coef(y, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y,y_pred)
print ("----------------------------///////------------------")
print ('The Mean Squared Error is %f' %mse)
print('The Root Mean Square Error is %f' %rmse)
print('The R2 is %f' %r2)
print('The Correlation is %f ' %corr)
print('------------------////////////////----------------')

plt.scatter(x, y, s=10)
plt.plot(x, y_pred, color='r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

```python
#Parabola with negative values 
import operator 
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures

np.random.seed(0)
x = np.arange(-5.0, 5.0, 0.1)
y = x**2 
y_noise = 2 * np.random.normal(size=x.size)
y = y + y_noise

# transforming the data to include another axis
x = x[:, np.newaxis]
y = y[:, np.newaxis]

polynomial_features= PolynomialFeatures(degree=2)
x_poly = polynomial_features.fit_transform(x)

model = LinearRegression()
model.fit(x_poly, y)
y_poly_pred = model.predict(x_poly)

corr = index_corr_coef(y, y_poly_pred)
rmse = np.sqrt(mean_squared_error(y,y_poly_pred))
mse = mean_squared_error(y,y_poly_pred)
r2 = r2_score(y,y_poly_pred)
print ("----------------------------///////------------------")
print ('The Mean Square Error is %f' %mse)
print('The Root Mean Square Error is %f' %rmse)
print('The R2 is %f' %r2)
print('The Correlation is %f ' %corr)
print ("----------------------------///////------------------")

plt.scatter(x, y, s=10)
# sort the values of x before line plot
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x,y_poly_pred), key=sort_axis)
x, y_poly_pred = zip(*sorted_zip)
plt.plot(x, y_poly_pred, color='m')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Varible')
plt.show()
```
